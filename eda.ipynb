```python
# paid/free courses in each subject based on level
ptl.figure(figsize=(10,6))
v6 = sns.barplot(x='subject',y='price',data = udemy ,hue = 'level',ci=False)

# count of paid/unpaid courses
price_list=udemy["is_paid"].unique()
price_count=udemy['is_paid'].value_counts().reset_index()
fig=px.bar(price_count, x='index', y='is_paid', text='is_paid', color='is_paid',
             title='count of courses paid and unpaid for',
             labels={'index':'paid/unpaid courses','is_paid':'count of paid/unpaid courses'})
fig.update_layout(showlegend=False, width=600)
fig.show()
```

Number of paid courses are 3368 and free courses are 310

```python
# price vs users
sns.scatterplot(data=udemy,x='price',y='num_subscribers')
```

```python
# subscription Vs paid
sns.barplot(x='is_paid', y='num_subscribers', data=udemy, ci=False)
```

Subscription is more for free courses rather than the paid ones

```python
# Subscriptions Vs price
price=pd.cut(x=udemy['price'], bins=[1, 50, 100, 150, 200])
sns.barplot(y=price, x='num_subscribers', data=udemy, ci=False)
```

Even though the number of subscribers opting for free courses was much higher than the paid ones, those who opted for the paid ones chose the course in range $150-200 the most expensive range.

```python

# Subscriptions Vs Duration
duration=pd.cut(x=udemy['content_duration'], bins=[1, 10, 20, 30, 40,50,60,70,80])
sns.barplot(x=duration, y='num_subscribers', data=udemy, ci=False)
```

Maximum number of subscribers lie in the range 30-60 hours range

```python
#Subscriptions Vs Reviews
sns.lineplot(y='num_reviews', x='num_subscribers', data=udemy, ci=False)
```

No particular pattern can be seen here

```python
# Most Popular courses
udemy[['course_title', 'num_subscribers','subject']].sort_values('num_subscribers', ascending = False).head(5)
```

```python
# Most engaging Courses
udemy['Engagement'] = udemy['num_subscribers'] + udemy['num_reviews']
udemy[['course_title', 'Engagement','subject']].sort_values('Engagement', ascending = False).head(5)
```

```python
# What courses offer the best cost benefit
# we can determine this by finding the course with the least amount paid, but has the highest number of subscribers. 
cost_benefit = udemy[(udemy['price']<=udemy['price'].mean()) & 
(udemy['num_subscribers']>=udemy['num_subscribers'].mean())].sort_values(('num_subscribers'),ascending=False)['course_title'].head(1).unique()
print("The course which offers the best cost benefit is : \n", cost_benefit)
```
